```python
import pandas as pd
import numpy as np
import pickle
```

```python
dataset = pd.read_csv('Data/Processed/training_output_csv.csv', header=None)
dataset.head()
```

# DATA PROCESSING

```python
dataset = dataset.drop(dataset.columns[0], axis=1)
dataset.head()
```

```python
X, y = dataset.iloc[: , 1:], dataset.iloc[: , 0]
```

```python
np.unique(y, return_counts=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=124)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

best_k = 1
best_acc = 0

for k in range(3, 50):
    model = KNeighborsClassifier(10)
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    if acc > best_acc: 
        best_k, best_acc = k, acc

print("Best K value: ",best_k)
```

```python
model = KNeighborsClassifier(1)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
filename = 'models/knn.sav'
pickle.dump(model, open(filename, 'wb'))
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
filename = 'models/random_forest.sav'
pickle.dump(model, open(filename, 'wb'))
```
